```python
# All data from CSV file COVID_19_time_series_clean_complete_All_Data 
file_to_load = "csv_files/COVID_19_time_series_clean_complete_All_Data.csv"
covid_19 = pd.read_csv(file_to_load)
confirmed_19_df_by_country = pd.DataFrame(covid_19)
del confirmed_19_df_by_country["Unnamed: 0"]
confirmed_19_df_by_country
```

```python
# To filter dataframe as of 04_30_2021
filt = confirmed_19_df_by_country["Date"] == "2021-04-30"
confirmed_19_df_by_country[filt]
```

```python
# This plot is for reference only and was not used for the final report package.
plt.figure(figsize=(40,20))
y_axis = confirmed_19_df_by_country[filt]["Confirmed"]
x_axis = confirmed_19_df_by_country[filt]["Country/Region"]
plt.bar(x_axis, y_axis, color='r', alpha=1.0, align="center")
plt.xlabel("Country")
plt.ylabel("Millions")
plt.title("Global - Confirmed as of 04_30_2021 (cummulative)")
plt.xticks(rotation = 90)
plt.show()
```

```python
# filter by date = ['2021-04-30']
filt = (confirmed_19_df_by_country["Date"] == "2021-04-30") & (confirmed_19_df_by_country["Confirmed"] >= 1000000)
confirmed_19_df_by_country[filt].sort_values(by = "Confirmed", ascending = False)
```

```python
# Globl Bar Chart as of 04/30/2021 with 1M and greater confirmed cases. 
plt.figure(figsize=(40,20))
y_axis = confirmed_19_df_by_country[filt].sort_values(by = "Confirmed", ascending = False)["Confirmed"]
x_axis = confirmed_19_df_by_country[filt].sort_values(by = "Confirmed", ascending = False)["Country/Region"]
pos = np.arange(len(x_axis))
plt.bar(x_axis, y_axis, width = 0.8, color='r', alpha=1.0, align="center")
plt.xlabel("Country", fontsize = 30)
plt.ylabel("Ten's of Millions", fontsize = 30)
plt.title("Global - Confirmed Covid 19 as of 04_30_2021 (cumulative, 1M or greater)", fontsize = 30)
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 20)

for index,data in enumerate(y_axis):
    plt.text(x=index-.3, y =data+200000, s=f"{format((data/10000000),'.2f')}", fontdict=dict(fontsize=20), weight="bold") 

plt.show()
```

```python
plt.savefig("global_confirmed_cummulative_by_country_1M_or_greater.png")
```

```python
# Bar Chart of Deaths / Confirmed Cases - Note: Mexico – usually high mortality rate at 9% in relation confirmed cases.
filt3 = (confirmed_19_df_by_country["Date"] == "2021-04-30") & (confirmed_19_df_by_country["Confirmed"] >= 2000000)
filt4 = confirmed_19_df_by_country[filt3]
filt4["Percentage"] = ((filt4["Deaths"]) / (filt4["Confirmed"]) * 100)
```

```python
filt5 = filt4.sort_values(by = "Percentage", ascending = False)
```

```python
# Mexico – note the usually high mortality rate at 9% in relation confirmed cases.
filt6 = filt5.sort_values(by = "Percentage", ascending = False)
y_axis = filt5["Percentage"]
x_axis = filt5["Country/Region"]
plt.figure(figsize=(40,20))
plt.bar(x_axis, y_axis, width = 0.8, color='r', alpha=1.0, align="center")
plt.xlabel("Country", fontsize = 30)
plt.ylabel("Percentage (Deaths/Confirmed)", fontsize = 30)
plt.title("Global - Covid 19 as of 04_30_2021 - Cummulative Deaths / Cummulative Confirmed (cumulative, sample of 2M or greater)", fontsize = 30)
plt.xticks(rotation = 45, fontsize = 20)
plt.yticks(fontsize = 30)


for index,data in enumerate(y_axis):
    plt.text(x=index-.2, y =data+.05, s=f"{format((data),'.1f')}" + "%", fontdict=dict(fontsize=20), weight="bold") 
    
plt.show()
```
