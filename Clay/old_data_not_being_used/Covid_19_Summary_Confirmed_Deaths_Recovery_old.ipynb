```python
import json
import requests 
import numpy as np
import requests
from scipy.stats import linregress
import pprint as pprint
import pandas as pd
import datetime
import matplotlib.pyplot as plt
from citipy import citipy
# from api_keys import weather_api_key
# from api_keys import g_key
import time as time
#import wget as wget
```

```python
file_to_load = 'covid_19_data.csv'
covid_19 = pd.read_csv(file_to_load)
confirmed_19_df = pd.DataFrame(covid_19)
# del confirmed_19_df["Unnamed: 0"]
confirmed_19_df
```

```python
del confirmed_19_df["SNo"]
del confirmed_19_df["Province/State"]
del confirmed_19_df["Last Update"]
confirmed_19_df.rename({"ObservationDate":"Date"}, axis=1, inplace=True)
confirmed_19_df
```

```python
# Date from sting to datetime
confirmed_19_df['Date'] = pd.to_datetime(confirmed_19_df['Date'])
confirmed_19_df.dtypes
```

Global By Country and Month

```python
confirmed_19_df_month_country = confirmed_19_df.set_index("Date").groupby("Country/Region").resample("M")
confirmed_19_df_month_country = confirmed_19_df_month_country.sum()
```

```python
confirmed_19_df_month_country.to_csv('COVID19_by_country_month.csv')
```
