```python
#dependencies
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#data
combined_df = pd.read_csv('combined_data.csv')
combined_df.head()
```

```python
#plot
ax = combined_df.plot(kind="scatter", x="percapita", y="full Vac by population", grid=True, figsize=(8,4),
              title="FullVac ratio to population percapita, Deaths(s), Vaccine(c)", s=combined_df["Deaths Ratio by population"]*100, c=combined_df["Vaccine Type Count"], colormap='jet')
ax.set_xscale('log')
plt.xlim([100,1000000])
plt.ylim([-5,80])
plt.show()
#plt.savefig("scatter_color_vacs.png")
```
