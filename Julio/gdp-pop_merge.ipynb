```python
# apps
import pandas as pd
import numpy as np
```

```python
# get data sets
path_gdp = "../Resources/gdp-csv.csv"
path_pop = "../Resources/WorldPopulation.csv"
path_locs = "../Resources/time_series_covid_19_deaths.csv"

gdp_raw_df = pd.read_csv(path_gdp, skiprows=4)
#gdp_raw_df.head(15)
pop_raw_df = pd.read_csv(path_pop)
#pop_raw_df.head(15)
#pop_raw_df.loc[pop_raw_df["Country Name"] == "Wallis"]
locs_raw_df = pd.read_csv(path_locs)
#locs_raw_df.head()
#locs_raw_df.loc[locs_raw_df["Country/Region"] == "France"]
#locs_raw_df.loc[locs_raw_df["Province/State"] == "Wallis and Futuna"]
```

```python
# get just desired columns
#gdp
gdp_gdp_2019 = []
gdp_country_2019 = []
gdp_code_2019 = []

counter = len(gdp_raw_df["2019"])
#print(counter)

for i in np.arange(counter):
    gdp = gdp_raw_df["2019"][i]
    gdp_gdp_2019.append(gdp)  
    country = gdp_raw_df["Country Name"][i]
    gdp_country_2019.append(country)
    code = gdp_raw_df["Country Code"][i]
    gdp_code_2019.append(code)

#gdp df
gdp_raw_2019_df = pd.DataFrame({"Country": gdp_country_2019,
                                "Code": gdp_code_2019,
                                "2019 GDP":gdp_gdp_2019
                                })

gdp_raw_2019_df.head(15)
```

```python
#get just desired columns
#pop
pop_pop_2019 = []
pop_country_2019 = []
pop_code_2019 = []

counter = len(pop_raw_df["2019"])
#print(counter)

for i in np.arange(counter):
    pop = pop_raw_df["2019"][i]
    pop_pop_2019.append(pop)  
    country = pop_raw_df["Country Name"][i]
    pop_country_2019.append(country)
    code = pop_raw_df["Country Code"][i]
    pop_code_2019.append(code)
    
pop_raw_2019_df = pd.DataFrame({"Country": pop_country_2019,
                                "Code": pop_code_2019,
                                "2019 Population":pop_pop_2019
                                })

pop_raw_2019_df.head(15)
```

```python
#get just desired columns
#locs
locs_lat_2019 = []
locs_lon_2019 = []
locs_country_2019 = []
locs_state_2019 = []
deaths_5121 = []

counter = len(locs_raw_df["Country/Region"])
#print(counter)

for i in np.arange(counter):
    lat = locs_raw_df["Lat"][i]
    locs_lat_2019.append(lat)  
    lon = locs_raw_df["Long"][i]
    locs_lon_2019.append(lon)
    country = locs_raw_df["Country/Region"][i]
    locs_country_2019.append(country)
    state = locs_raw_df["Province/State"][i]
    locs_state_2019.append(state)
    deaths = locs_raw_df["5/1/21"][i]
    deaths_5121.append(deaths)
    
locs_raw_5121_df = pd.DataFrame({"Country": locs_country_2019,
                                "State": locs_state_2019,
                                "Latitude":locs_lat_2019,
                                 "Longitude":locs_lon_2019,
                                 "Deaths to 5/1/21": deaths_5121
                                })

locs_raw_5121_df.head()
locs_clean_5121_df = locs_raw_5121_df.copy()
#locs_clean_5121_df.size
```

```python
#clean location and deaths
#australia
australia = locs_raw_5121_df.loc[locs_raw_5121_df["Country"] == "Australia"]
australia_lat = australia["Latitude"].mean()
#print(australia_lat)
australia_lon = australia["Longitude"].mean()
#print(australia_lon)
australia_deaths = australia["Deaths to 5/1/21"].sum()
#print(australia_deaths)
#print(australia)
locs_clean_5121_df.drop(locs_clean_5121_df.loc[9:15].index, inplace=True)
locs_clean_5121_df.loc[8,"Latitude"] = australia_lat
locs_clean_5121_df.loc[8,"Longitude"] = australia_lon
locs_clean_5121_df.loc[8,"Deaths to 5/1/21"] = australia_deaths
#canada
canada = locs_raw_5121_df.loc[locs_raw_5121_df["Country"] == "Canada"]
canada_lat = canada["Latitude"].mean()
#print(canada_lat)
canada_lon = canada["Longitude"].mean()
#print(canada_lon)
canada_deaths = canada["Deaths to 5/1/21"].sum()
#print(canada_deaths)
locs_clean_5121_df.drop(locs_clean_5121_df.loc[40:54].index, inplace=True)
locs_clean_5121_df.loc[39,"Latitude"] = canada_lat
locs_clean_5121_df.loc[39,"Longitude"] = canada_lon
locs_clean_5121_df.loc[39,"Deaths to 5/1/21"] = canada_deaths
#china
locs_raw_5121_df.loc[70,"Country"] = "Hong Kong"
locs_clean_5121_df.loc[70,"Country"] = "Hong Kong"
#
china = locs_raw_5121_df.loc[locs_raw_5121_df["Country"] == "China"]
china_lat = china["Latitude"].mean()
print(china_lat)
china_lon = china["Longitude"].mean()
print(china_lon)
china_deaths = china["Deaths to 5/1/21"].sum()
print(china_deaths)
locs_clean_5121_df.drop(locs_clean_5121_df.loc[59:69].index, inplace=True)
locs_clean_5121_df.drop(locs_clean_5121_df.loc[71:91].index, inplace=True)
locs_clean_5121_df.loc[58,"Latitude"] = china_lat
locs_clean_5121_df.loc[58,"Longitude"] = china_lon
locs_clean_5121_df.loc[58,"Deaths to 5/1/21"] = china_deaths
locs_clean_5121_df.loc[locs_clean_5121_df["Country"] == "China"]
#locs_raw_5121_df.loc[locs_raw_5121_df["Country"] == "China"]
#locs_raw_5121_df.loc[70]
```

```python
#list of countries for vaccine sheet
path_vacs = "../Resources/country_vaccinations.csv"

vacs_raw_df = pd.read_csv(path_vacs)
#vacs_raw_df.head(15)

#just lines for 5/1/2021
vacs_raw_512021_df = vacs_raw_df.loc[vacs_raw_df['date']== '5/1/2021']
#vacs_raw_512021_df.head(5)
#vacs_raw_df.loc[vacs_raw_df['country'] == 'Yemen'] ------add
#vacs_raw_df.loc[vacs_raw_df['country'] == 'Wallis and Futuna']
```

```python
#just needed columns for vaccine list
#vacs
vacs_country_5121 = []
vacs_code_5121 = []
vacs_total_5121 =[]
vacs_full_5121 = []
vacs_vacs_5121 = []

counter = len(vacs_raw_512021_df["country"])
#print(counter)
vacs_5121_index = vacs_raw_512021_df.index.tolist()

for i in vacs_5121_index:
    country = vacs_raw_512021_df["country"][i]
    vacs_country_5121.append(country)  
    code = vacs_raw_512021_df["iso_code"][i]
    vacs_code_5121.append(code)
    total = vacs_raw_512021_df["total_vaccinations"][i]
    vacs_total_5121.append(total)
    full = vacs_raw_512021_df["people_fully_vaccinated"][i]
    vacs_full_5121.append(full)
    vacs = vacs_raw_512021_df["vaccines"][i]
    vacs_vacs_5121.append(vacs)
    
#print(vacs_vacs_5121)

vacs_raw_5121_df = pd.DataFrame({"Country": vacs_country_5121,
                                "Code": vacs_code_5121,
                                "Total Vaccinations": vacs_total_5121,
                                "Total Fully Vaccinated": vacs_full_5121,
                                "Vaccinations Used": vacs_vacs_5121
                                })
#vacs_raw_5121_df.head(20)
```

```python
#split vaccine types
#get list of vaccines
vac_type_list = []
counts_list = []
counts_list_entry = []
country_entry = []

counter = len(vacs_raw_5121_df["Vaccinations Used"])
#print(counter)

#separate vaccines and list
for i in np.arange(counter):
    first = vacs_raw_5121_df["Vaccinations Used"][i]
    first_l = first.strip('()').split(',')
    country = vacs_raw_5121_df["Country"][i]
    counts= len(first_l)
    counts_list.append(counts)
    for j in np.arange(counts):
        vac_type_list.append(first_l[j])
        country_entry.append(country)
        counts_list_entry.append(counts)
        
#vac_type_list
#counts_list_entry
#country_entry
#len(vac_type_list)
#len(country_entry)
#len(counts_list_entry)

vac_type_used = pd.DataFrame({"Country": country_entry,
                             "Vaccine Used": vac_type_list,
                             "Number of Vaccines Used": counts_list_entry
                             })

vac_type_used["Vaccine Used"]=vac_type_used["Vaccine Used"].str.replace(' ', '')

vac_type_list1 =[]
vacs = vac_type_used["Vaccine Used"].unique()
vac_type_list1 = vacs.tolist()
vac_type_list1
#vac_type_used
```
